#### Define Functions to Retrieve Image File info

```python
def getFilepath(filename):
    # filepath = '/'.join(HOpreds.filename[0].split('/')[5:])
    return '/'.join(filename.split('/')[5:])
```

```python
def getFileCategory(filename):
    # filepath = '/'.join(HOpreds.filename[0].split('/')[5:])
    return filename.split('/')[-2]
```

```python
def getFileID(filename):
    # fileID = HOpreds.filename[0].split('/')[-1][:-4]
    return filename.split('/')[-1][:-4]
```

#### Define Functions to Derive bounding box of image wrt geospatial info --> geometry for geopandas?

```python
# http://nbviewer.jupyter.org/github/mqlaql/geospatial-data/blob/master/Geospatial-Data-with-Python.ipynb

import gdal

def GetImgTransform(gtif):
    #arr = gtif.ReadAsArray()
    tif_transform = gtif.GetGeoTransform() # ulx, xres, xskew, uly, yskew, yres
    #print(tif_transform)
        
    #http://www.gdal.org/gdal_datamodel.html
    ulx, xres, xskew, uly, yskew, yres = tif_transform
    lrx = ulx + (gtif.RasterXSize * xres) #RasterXSize = 255
    lry = uly + (gtif.RasterYSize * yres) #RasterYSize = 250
    bbox = (ulx, uly, lrx, lry) #diagonals
    #print(boundingbox)

    return bbox

def readGeotiff_GetImgTransformNgeomPoly(filename):
    #gtif1 = gdal.Open(path+'building_set/buildings/100208.tif')
    gtif = gdal.Open(filename)
    
    bbox = GetImgTransform(gtif)
    
    bbox_poly = shapely.geometry.box(*bbox)
    
    return bbox_poly
```

```python
def getGEOinfo(FileList):

    geoInfo=[]

    cnt=0
    for f in FileList:
        #print(f)

        #filepath = getFilepath(f)
        #print(filepath)

        fileCAT = getFileCategory(f)
        #print(fileCAT)

        fileID = getFileID(f)
        #print(fileID)

        bbox_poly = readGeotiff_GetImgTransformNgeomPoly(f)
        #print(bbox_poly)

        geoInfo.append([fileCAT, fileID, bbox_poly])

        cnt +=1
        if cnt>0 and cnt%100==0:
            print('retrieved ' + format(cnt) + ' geoInfo ----------')
    
    return geoInfo
```

#### Get Geometry for Building Images

```python
geoInfo = getGEOinfo(Build_dir)
```

```python
buildGPD = gpd.GeoDataFrame(geoInfo, columns=['class_name','feature_id','geometry'])
buildGPD
```

```python
buildGPD.plot()
```
